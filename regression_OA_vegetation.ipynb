## 1. Background

Earth's population is growing (**[United Nations, 2022](#references)**) and increasingly more people are moving to cities in search of better job opportunities, healthcare, education, and quality of life (**[United Nations, 2018](#references)**). Urbanization puts pressure on urban planners to, on the one hand, address the increased demand on housing (along with the need for additional commercial, bussiness, and educational buildings) while, on the other hand, maintain good and sustainable urban living conditions. Urban densification is often considered as a more preferable solution compared to urban expantion as it allows more people to use the already existent infrastructure while valuable agricultural land frequently surrounding urban suburbs is preserved (**[Pelczynski & Tomkowicz, 2019](#references)**). Nevertheless, urban densifications are not problem-free. One of the most challenging problems urban planners are called to solve when planning an urban densification is to ensure adequate access to daylight/sunlight. 

*Daylight* is described as the infrared, visible and ultraviolet part of the global solar radiation (direct or indirect sunlight:  reflected or diffused) that enters the room of a building through a window or other kind of opening (**[EN 12665:2018, 2018, 3.4.7](#references)**). *Sunlight* (infrared, visual, and ultraviolet light) is the light of the sun that enters a room directly without previously being reflected. Insufficient or limited availability of natural daylight has negatively been associated with higher energy demands for lighting (**[Lechner, 2014](#references)**), poor health (**[Lam & Ripman, 1992](#references)**; **[Osterhaus, 2005](#references)**; **[Lee et al., 2022](#references)**), and reduced productivity in both work and educational environments (**[Vischer, 2008](#references)**). Therefore, it is clear how important adequate access to natural light is for urban populations. But how do we assure that this is achieved?

Existing national and international regulations and guidelines concerning daylight are there to aid urban planners and real estate developers in making well-informed choices. A large number of metrics has been developed to measure the magnitude and quality of daylight (**[Ayoub, 2019](#references)**). These metrics have different requirements on input data and, depending on their complexity, are suitable for different stages of the urban planning process. The intricate nature of contemporary urban environments in combination with the diverse number and complexity of daylight evaluation methods (metrics) hampers their task. This necessitates the implementation of systematic analyses that encompass broader geographical areas. As a result, a requirement for a standardized approach to integrating daylight simulations into the planning process (Fig. 1) is identified.

In this study, the focus is set on daylight metrics that can be utilised at early stages of the urban planning process (e.g., detailed development plan stage), when little to no information is available about a planned building. *Detailed development plans* (DDP) are legally binding documents that dictate the rules regarding how a certain parcel of urban land (along with its resources; e.g. soil, water, etc.) may be exploited (Fig. 1). Some of the most common rules that are defined in such documents are i.a., the exact part of the area that can be built/exploited as well as the maximum permitted height of buildings within that area. For this purpose, urban planners perform various analyses (either alone or in cooperation with consultants; e.g. daylight or noise experts) to determine the threshold values to those rules. In order to achieve this, they often use simple 3D box models to represent buildings and test what combination of building typologies and heights ensures the best results. There is a standard practice in Sweden to use the *Obstruction Angle* daylight metric when creating a detailed development plan, as a first attempt of evaluating daylight access in rooms. Typically, this is done for the ground floor, whose rooms are more affected by surrounding tall buildings. *Obstruction angle* (OA) is defined as the angle between the horizontal plane and a line connecting, e.g., the center point of a window and the highest point of an obstruction (Fig. 2).

<br>
<center><font size="-1"><b>Figure 2</b>: Obstruction Angle (OA) computation example <br> <i>(Source: Pantazatou et al., 2023, p.24)</i></font></center>

The OA is computed to obtain an idea of how likely it is for future planned buildings to comply to the legislated daylight requirement (i.e., a $DF > 1%$ must be obtained by any room that is visited more than frequently according to *section 6:322* of the *building regulations* (BBR) by the *National Board of Housing, Building and Planning*; **[Boverket, 2011](#references)**). For more information on the Daylight Factor (DF), go to [Appendix 1](#appendix1). It has been found that complying to the aforementioned legislated requirement is equivalent to: 

- Obstruction angle as: $OA <= 30^{\circ}$ 
- Window-to-floor area ratio as: $A_{window} \ge f_{(x)} × A_{floor}$ <br> 

where:

$A_{window}$ = total window area of a room <br>
$A_{floor}$ = total floor area of a room <br>
$f_{(x)}=\begin{cases}
0.075, & \text{ x } \in [0, 10],\\
0.0025 × x + 0.05, & \text{ x } \in [10, 30]
\end{cases} \text{    (for x = window obstruction angle)} $

The choice of daylight metric dictates the choice of software, which, in turn, decides the type of input data. The OA requires only information on building geometries (window information must be provided). Many municipalities in Sweden have produced 3D representations of their urban areas. In general, these 3D models were mainly used for visualization purposes. They have been developed  gradually (since the early 2000), and in the absence of national guidelines on following a specific standard, it is possible that they differ significantly in the way they were created. As the OA may depend on the level of detail buildnings/building parts (e.g., roofs, balconies) are represented in (**[Pantazatou et al., 2023](#references)**), it is important that we use a 3D city model standard to set the minimum requirements for computing a certain daylight metric. This would assure that all daylight metric computations are based on the same assumptions and the accuracy of the produced results is relatable.

3D city models have the potential to be utilized as input data for daylight simulations. Nevertheless, to achieve this, it is necessary that the 3D city models encompass all necessary information needed for computing daylight metrics. Additionally, the level of detail as well as the quality of the corresponding data must meet adequate standards. *CityGML* is a semantic 3D city model standard approved by the *Open Geospatial Consortium* (OGC). *Semantic 3D city models* are 3D digital  representations of the urban environment, including buildings, streets, trees, bridges, and the terrain (**[Gröger et al., 2012](#references)**). Besides information on the geometry of every urban object, these models have also the capacity to store semantic information for every geometry (e.g., year a building/building part was built) (**[Gröger et al., 2012](#references)**). Besides buildings, CityGML also includes a model dedicated to the 3D representation of vegetation. This model makes a distinction between solitary vegetation objects (e.g., trees) and vegetated areas (e.g., forests, grasslands/parks). Vegetation objects from the first category are modelled by the class *SolitaryVegetationObject*, while vegetated areas covered by specific vegetation are modelled by the class *PlantCover MultiSolid* (3D: e.g., forests) or *Plantcover MultiSurface* (2D: e.g., grassland) (**[Gröger et al., 2012](#references)**) (Fig. 3).

In order to maintain the sensitive ballance of the biodiversity of urban ecosystems, reducing/counteracting/limiting the effects of urban heat island, and continue contributing to the wellbeing of urban populations, it is essential to ensure that enough vegetated areas are present especially in densely built urban areas (**[Anderson & Gough, 2021; Endsley, 2018; Khan et al., 2021](#references)**). Though trees may have a very possitve influence on a variety of aspects of urban life, when it comes to daylight, this is not always the case. Trees with dense foliage (e.g., English oak) have been found to prevent up to 80% of the incoming sunlight in full leaf and up to 55% during winter when there is no foliage (**[Littlefair, 2011](#references)**), while trees with less dense foliage have also been found to prevent incoming sunlight enough to constitute their inclusion in daylight modelling highly relevant (**[Mashaly et al., 2015](#references)**). Therefore, it remains to be invastigated if and to what extent trees located close to buildings affect the access to daylight. This is important to examine both from an indoor lighting perspective as well as from an energy perspective (i.e., potential to install photovoltaic solar panels (PVS) on building roofs or facades). If vegetation is found to have a significant limiting impact on indoor lighting or the potential of installing PVSs, it would be essential to include vegetation in the daylight simulations executed at early stages of the urban planning process. Particularly so, when planning urban densifications. It should be mentioned that, up until recently, it was not standard practice to include vegetation in daylight simulations, mainly because trees were not considered as a permanent obstruction of daylight in an urban settings due to seasonal changes in their daylight-blocking capability as well as because they could be cut. However, in recent years, the development of more complex metrics as well as the densification of cities has lead several researchers to start studying the limiting effects of vegetation in relation to daylight access (**[Ayoub, 2019](#references)**).

## 2. Aim
The overarching objective is to guarantee that the upcoming Swedish standard for 3D city models (currently under development) 3CIM (**[Uggla et al., 2023](#references)**) facilitates the implementation of daylight simulations during the initial phases of urban densification projects. To this end, we have identified daylight metrics that are explicitly mentioned or implicitly inferred to by Swedish or European daylight-regulating legislation or recommendations (**[Pantazatou et al., 2023](#references)**). The OA is one of these metrics. 

The particular aim of this study is to determine whether vegetation should be included in 3CIM, so that it can facilitate the execution of daylight simulations at different stages of the urban planning process. For this purpose, this notebook is dedicated on presenting the post-processing of OA daylight simulation output executed over a 3CIM test model including vegetation (trees). More in particular, the notebook presents correlation and regression analysis results of the obstruction angle (OA) simulation output in connection to vegetation and other factors believed to affect the daylight access of rooms in densely built urban environments. The scenario the csv-file corresponds to is the "existing buildings with vegetation". Planned buildings were disregarded in this case.

## 4. Data & methods
The following datasets were used in the context of this study:

- 3D city model with buildings (3CIM - experimental version)
- 3D window information (**[Hartman, 2021](#references)**)
- 2D point location of trees (Malmö municipality)

<br>

The 3D tree information was created according to the following. An inventory of the height and crown width of every tree was made, as part of this study, based on Google Earth images. This inventory was linked with Malmö municipality’s 2D point tree layer and inserted into an [FME](https://www.safe.com/products/)-script to produce CityGML SolitaryVegetationObjects using a program from [con Terra GmbH & Virtual City Systems (2021)](https://hub.safe.com/publishers/con-terra?page=1&page_size=10). The [FME-script](https://hub.safe.com/publishers/con-terra/templates/citygml-create-solitaryvegetationobject-with-implicit-representation) included two SketchUp prototype models of deciduous and coniferous trees. It required the geographic coordinates, (latitude, longitude, and elevation), height, crown width, and type of every tree in the study area for adjusting the prototypes to those measures and produce a corresponding 3D representation of the tree. In addition to the 3CIM data we also used 3D window information. This information was produced as part of a collaboration between researchers at the Norwegian University of Science and Technology and Lund University (**[Hartman, 2021](#references)**). 


A tool was developed in [ArcPy](https://pro.arcgis.com/en/pro-app/latest/arcpy/get-started/what-is-arcpy-.htm) to compute the general case of the OA of a rectangular window. The documentation of the tool along with the corresponding code are availbale on a separate repository on [GitHub](https://github.com/Sustainable3DCities/Obstruction_Angle_Tool). It requires 3 input data sets (Fig. 4):

The DSM was created by resampling the DTM to 0.1m resolution and adding the rasterized 3D building and vegetation information. The 2D building footprints were produced by getting the footprint of the corresponding 3D buiding objects. These operations were performed in ArcGIS Pro. A column was added to the window and 2D building footprint layers to store the ID of the building they correspond to. 

Once all input datasets are prepared, the OA tool was executed for different scenarios. The scenario presented in this notebook is the following "existing buildings with vegetation". The OA tool first computes the centroid and orientation of every window. Next, a viewshed analysis is carried out over the DSM raster for every centroi to identify which cells are seen within a horizontal scan range from 85—95° (vertical distance to obstruction). From the two produced viewshed rasters, the one that is empty or overlaps with the building footprint is deleted. The cells of the remaining viewshed raster are converted to points and the OA is calculated for all points. The max OA is added as a result in the corresponding column of the window layer’s attribute table.

Once the OA have been calculated, we add columns to the window centroid layer to store, the type, height and crown width of the tree that is obstructing daylight access to a particular window (not all windows are obstructed by trees). We also add columns for storing the floor and orientation (i.e., North, Northeast, East, Southeast, South, Southwest, West, Northwest) of every window. Finally, we add columns to store the 2D and 3D Euclidean distance between the centroid and the closest tree or the tree on which the obstruction point is located. The trees are represented either as a 3D point or as a 3D tree object. Once all the fields have been filled with values, the layer is exported and imported to this notebook. 

Next, we perform a regression analysis over the aforementioned variables and the windows whose OA>30.

The dependent variable is the **OA** while the independent variables are:
- tree height
- crown width of tree
- tree type (e.g., coniferous or deciduous)
- building floor the window is placed on
- window orientation
- 2D Euclidean distance between window centroid & its closest tree (tree represented as point)
- 2D Euclidean distance between window centroid & its closest tree (tree represented as 3D object)
- 3D Euclidean distance between window centroid & its closest tree (tree represented as 3D object)
- 2D Euclidean distance between window centroid & tree where the obstruction point is placed (tree represented as point)
- 2D Euclidean distance between window centroid & tree where the obstruction point is placed (tree represented as 3D object)
- 3D Euclidean distance between window centroid & tree where the obstruction point is placed (tree represented as 3D object)


We also perform a Pearson correlation analysis per floor over the same variables.

```python
def get_linear_Reg_coef(X, Y):
    
    #Create an instance of the class LinearRegression, which will represent the regression model:
    model = LinearRegression().fit(X, Y)
    
    #Coefficient of determination R2:
    r_sq = model.score(X, Y)
    
    #Get intercept value:
    intercept = round(model.intercept_,2)
    
    #Get model coefficient:
    m_coeff = round(model.coef_[0] ,2)
    
    #Get length of dependent variable:
    y_lenght = len(Y)
    
    #Return values:
    print("Model: ", model)    
    print("R\u00b2 : ", round(r_sq,2))
    print("Intercept : ", intercept)
    print("Model coefficient : ", m_coeff)
    
    
    #Return values:
    return model, r_sq, intercept, m_coeff, y_lenght, 
```

## 6. Discussion
From the regression analysis we find that the following variables have a statistically significant impact on limiting daylight access for windows that exceed OA>30:

- 3D Euclidean distance between window centroid and tree where the obstruction point is found (tree is represented as 3D object) **negative correlation**
- building floor where the window is located  **negative correlation**
- orientation of window **possitive correlation** 
- tree height **possitive correlation**
- crown width of tree **possitive correlation**

The regression model has a high confidence as it describes **92%** (Adjusted R-squared) of the cases. Like expected, as the 3D distance between a window centroid and the tree on which the obstruction point is located is increasing the OA is decreasing. This is also the case for the buildings floors, windows placed in rooms of lower floors have higher OA. As we assigned integer numbers clockwise starting from *North* for the orientation, we find that as the orientation numbers are increasing (i.e., going to South and Southwest facing facades), the obstruction angle is increasing. This is expected as more rooms here in the north as placed to face the South in order to increase access to daylight. This may also be a factor affecting the planting and growth of urban trees. Also, the height and crown width of trees were found to have possitive correlation to OA. The taller or wider the tree, the higher the OA. Most notably, an increase by 1 floor can lead to a decrease in the OA that is up to 9 degrees. Additionally, an increase of 1m in the 3D distance between a window centroid and the location on a tree where the obstruction point is placed, can decrease the OA by almost 7 degrees.

Once executing the Pearson correlation method over all floors for the aforementioned variables, it turns out that only the building *floor* has a relatively strong possitive correlation with the likelihood of passing the 30-degree daylight-criterion. At the same time, the height and crown width of a tree are found to have a very week possitive correlation. The higher the floor the higher the Pearson correlation output for complying to the 0-degree daylight-criterion.

## Appendix 1

### A1.1. Daylight factor (DF)
The DF (Fig. A1.1) is a daylight metric that corresponds to a percentage expressed as the ratio of the illuminance in a room compared to the illuminance on a horizontal plane under an unobstructed hemisphere with an overcast sky (**[Dubois et al., 2019](#references)**). 

<img src="https://github.com/KarolinaPntzt/regression/blob/main/img/daylight_factor.png?raw=true" width=60%></img>

It should be noted that the DF is only calculated for a single point based on diffuse light and does, therefore, not take into consideration orientation. Daylighting is very much climate-, time- and orientation-dependent. The metric’s failure to account for how climate-, latitude- and orientation-related variations affect the estimated daylight conditions, is rather a significant deficit (**[Mardaljevic et al., 2009](#references)**). Amongst the advantages of this metric is that its results can easily be described and understood by a design team (**[Reinhart et al., 2006](#references)**). Moreover, it expresses the lighting potential of a room and its windows as well as giving a good indication of the contrast between indoor and outdoor lighting conditions (**[Love & Navvab, 1994](#references)**).

<br>

### A1.2. Point daylight factor (DFp)
DFp is described in the Swedish Building Regulations (BBR), where it is calculated as a single DF-measurement taken at a  single timepoint at a point located 1m away from the darkest wall, in the middle of the room’s depth at 0.8m above the floor’s surface (Fig. A1.2).
